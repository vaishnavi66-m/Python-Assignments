```python
class Student:
    
    def set_values(self, name, age):
        
        self.name = name

        if age > 0:
            self.age = age

        else:
            print("Invalid Age! Age must be greater than 0")

    def get_values(self):
        print("Name:", self.name)
        
        if hasattr(self, 'age'):
            print("Age:", self.age)


# Object
s1 = Student()

s1.set_values("Vaishnavi", 21)
s1.get_values()
```

```python
class Student:

    def __init__(self):
        self._age = 0

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):

        if value > 0:
            self._age = value

        else:
            print("Invalid Age! Age must be greater than 0")


# Object
s1 = Student()

s1.age = 22

print("Age:", s1.age)
```

```python
class BankAccount:

    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):

        self.balance += amount

        print("Deposited:", amount)
        print("Current Balance:", self.balance)

    def withdraw(self, amount):

        if amount <= self.balance:

            self.balance -= amount

            print("Withdrawn:", amount)
            print("Current Balance:", self.balance)

        else:
            print("Insufficient Balance")


# Object
acc1 = BankAccount("Vaishnavi", 5000)

acc1.deposit(2000)

acc1.withdraw(3000)

acc1.withdraw(10000)
```

```python
class PasswordSystem:

    def __init__(self):
        self._password = ""

    @property
    def password(self):
        return self._password

    @password.setter
    def password(self, value):

        has_number = False

        for ch in value:
            
            if ch.isdigit():
                has_number = True

        if len(value) >= 8 and has_number:
            self._password = value
            print("Password Set Successfully")

        else:
            print("Invalid Password")
            print("Password must:")
            print("- Be at least 8 characters")
            print("- Contain at least one number")


# Object
p1 = PasswordSystem()

p1.password = "Python123"

print("Password:", p1.password)
```

```python
from abc import ABC, abstractmethod

# Abstract Class
class Animal(ABC):

    # Abstract Method
    @abstractmethod
    def sound(self):
        pass

    # Non-Abstract Method
    def display(self):
        print("This is a non-abstract method from Animal class")
```

```python
class Dog(Animal):

    # Implementing abstract method
    def sound(self):
        print("Dog barks")
```

```python
# Object of child class
d1 = Dog()

# Calling inherited non-abstract method
d1.display()
```

```python
# Calling implemented abstract method
d1.sound()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass


class Car(Vehicle):

    def start(self):
        print("Car started")

    def stop(self):
        print("Car stopped")


# Object
c1 = Car()

c1.start()
c1.stop()
```

```python
from abc import ABC, abstractmethod

# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    # Non-abstract method
    def display(self):
        print("Calculating Area")


# Circle Class
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print("Area of Circle:", 3.14 * self.radius * self.radius)


# Rectangle Class
class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print("Area of Rectangle:", self.length * self.width)
```

```python
class Parent:

    def __init__(self, name):

        self.name = name

        print("Parent Constructor Called")


class Child(Parent):

    def __init__(self, name, age):

        # Calling parent constructor
        super().__init__(name)

        self.age = age

        print("Child Constructor Called")

    def display(self):

        print("Name:", self.name)
        print("Age:", self.age)


# Object
c1 = Child("Vaishnavi", 21)

c1.display()
```

```python
class Demo:

    def __init__(self):

        # Public variable
        self.public_var = "Public"

        # Protected variable
        self._protected_var = "Protected"

        # Private variable
        self.__private_var = "Private"


# Object
d1 = Demo()

# Public Access
print("Public:", d1.public_var)

# Protected Access
print("Protected:", d1._protected_var)

# Private Access (will give error if accessed directly)
# print(d1.__private_var)

# Accessing private variable using name mangling
print("Private:", d1._Demo__private_var)
```

```python
def calculate_bill(discount=0, *items, **extra_items):

    total = sum(items)

    # Add extra items
    for value in extra_items.values():

        total += value

    # Apply discount
    final_amount = total - discount

    print("Total Amount:", total)

    print("Discount:", discount)

    print("Final Amount:", final_amount)


# Function Call
calculate_bill(
    50,
    100, 200, 300,
    delivery=40,
    packing=20
)
```

```python
# Custom Exception
class InsufficientBalanceError(Exception):
    pass


class BankAccount:

    def __init__(self, balance):

        self.balance = balance

    def withdraw(self, amount):

        if amount > self.balance:

            raise InsufficientBalanceError("Insufficient Balance")

        else:

            self.balance -= amount

            print("Withdrawal Successful")
            print("Remaining Balance:", self.balance)


# Object
acc = BankAccount(5000)

try:

    acc.withdraw(7000)

except InsufficientBalanceError as e:

    print(e)
```
