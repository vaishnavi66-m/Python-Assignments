```python
# Global variable
value = "Global Variable"

def variable_scope():
    
    # Local variable with same name
    value = "Local Variable"
    
    print("Inside function (Local):", value)
    
    # Accessing global variable using globals()
    print("Inside function (Global):", globals()['value'])

variable_scope()

print("Outside function:", value)
```

```python
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print("Palindrome Number")

else:
    print("Not a Palindrome Number")
```

```python
def find_duplicates(numbers):
    
    duplicates = []

    for i in numbers:
        
        if numbers.count(i) > 1 and i not in duplicates:
            duplicates.append(i)

    print("Duplicate Elements:", duplicates)

list1 = [1, 2, 3, 2, 4, 5, 1]

find_duplicates(list1)
```

```python
class A:
    
    def __init__(self):
        print("Constructor of Class A")

    # Instance Variable
    name = "Class A Variable"

    # Specific Methods
    def method_a1(self):
        print("Method A1 from Class A")

    def method_a2(self):
        print("Method A2 from Class A")

    # Common Method
    def common_method(self):
        print("Common Method from Class A")


class B(A):

    def __init__(self):
        super().__init__()
        print("Constructor of Class B")

    # Instance Variable
    name = "Class B Variable"

    # Specific Methods
    def method_b1(self):
        print("Method B1 from Class B")

    def method_b2(self):
        print("Method B2 from Class B")

    # Overridden Method
    def common_method(self):
        super().common_method()
        print("Common Method from Class B")


class C(B):

    def __init__(self):
        super().__init__()
        print("Constructor of Class C")

    # Instance Variable
    name = "Class C Variable"

    # Specific Methods
    def method_c1(self):
        print("Method C1 from Class C")

    def method_c2(self):
        print("Method C2 from Class C")

    # Overridden Method
    def common_method(self):
        super().common_method()
        print("Common Method from Class C")
```

```python
# Object of A
obj_a = A()

print("\n--- Methods of A ---")
obj_a.method_a1()
obj_a.method_a2()
obj_a.common_method()


# Object of B
obj_b = B()

print("\n--- Methods of B ---")
obj_b.method_a1()
obj_b.method_a2()
obj_b.method_b1()
obj_b.method_b2()
obj_b.common_method()


# Object of C
obj_c = C()

print("\n--- Methods of C ---")
obj_c.method_a1()
obj_c.method_a2()
obj_c.method_b1()
obj_c.method_b2()
obj_c.method_c1()
obj_c.method_c2()
obj_c.common_method()
```

```python
# Reference of A holding object of B
ref = B()

print("\nUsing superclass reference with B object:")
ref.common_method()


# Reference of A holding object of C
ref = C()

print("\nUsing superclass reference with C object:")
ref.common_method()
```

```python
class Vehicle:

    def start_engine(self):
        print("Vehicle engine started")


class Car(Vehicle):

    def start_engine(self):
        super().start_engine()
        print("Car engine started")


class ElectricCar(Car):

    def start_engine(self):
        super().start_engine()
        print("Electric car battery system started")


# Objects
v = Vehicle()
c = Car()
e = ElectricCar()

print("\nVehicle:")
v.start_engine()

print("\nCar:")
c.start_engine()

print("\nElectric Car:")
e.start_engine()
```
